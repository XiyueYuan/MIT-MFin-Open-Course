1. Interest 
- Compounding: 
let $r$ be interest rate, $n$ be years, $PV$ be present value, $FV$ be future value

$$FV = PV \times (1 + r)^n$$
$$PV = \frac{FV}{(1 + r)^n}$$

- Periodic Compounding
let $r$ be interest rate, $m$ be compounding frequency

$$PV = (1 + \frac{r}{m})^{mn}$$

$$ Let ~ k = \frac{r}{m}$$

$$ \lim\limits_{m\to \infty} (1 + \frac{r}{m})^{mn} = e^{rn}$$

Therefore, 
$$FV = PV \times e^{rn}$$

2. Discounting

$$PV = (1 + \frac{r}{m})^{mn}$$
Here, $r$ is discount rate: $r = ~ risk ~ premium + r_f$
- Discounting Factor: 
$$\frac{1}{1 + r}$$
$$PV = \sum c \times e^{-rn} = \int_{0}^{T}c \times e^{-rn} = \frac{c}{r} \times (1 - e^{-rT})$$


### Examples
- Scenario1: Discrete

You buy a 2-year Apple (AAPL) bond with a Face Value (FV) of $1,000. 

The Coupon Rate is 5%, paid annually. 

The market's Required Yield (r) is 4%


$$
PV = 50 \times \frac{1}{(1 + 0.04)^1} + 1050 \times \frac{1}{(1 + 0.04)^2} = 1018.86
$$

You buy a 2-year Apple (AAPL) bond with a Face Value (FV) of $1,000. 

The Coupon Rate is 4%, paid annually. 

The market's Required Yield (r) is 6%

$$
PV = 40 \times \frac{1}{(1 + 0.06)^1} + 1040 \times \frac{1}{(1 + 0.06)^2} = 963.34
$$


Coupon < Yield: Discount

Coupon = Yield: At Par

Coupon > Yield: Premium 



- Scenario2: Continuous

You deposit 10,000 USDC into a liquidity pool.

The protocol offers a Continuous APY (r) of 12%. 

You plan to leave the money there for 6 months (T = 0.5 years).

$$
FV = PV \times e^{rT} = 10000 \times e^{0.12 * 0.5} = 10618.37
$$

Imagine a DAO is "streaming" you a grant of $12,000 per year (C=12,000), distributed block-by-block.

To find the value of the next 3 months (0.25 years) of that stream today:

$$
PV = \frac{12000}{0.12} \times (1 - e^{-0.12 * 0.25}) = 2955.45
$$
